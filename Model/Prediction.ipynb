```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import keras
from datetime import date
from keras.models import load_model
from sklearn.preprocessing import MinMaxScaler
import yfinance as yf
```

```python
today = date.today()
years = 1
start_date = (today - (date(today.year + years, 1, 1) - date(today.year, 1, 1)))
end_date = date.today()

MMS = MinMaxScaler()
```

```python
def predict(data, n_days=10):

    model = load_model('Stock Prediction Model1.keras')
    df = data.copy()
    df[df.columns] = MMS.fit_transform(df[df.columns])

    sequences = []
    sequences.append(df[df.columns].iloc[-20:])
    x = np.array(sequences)

    df = df._append(pd.DataFrame(columns=df.columns,index=pd.date_range(start=df.index[-1], periods=n_days+1, freq='D')))
    upcoming_prediction = pd.DataFrame(columns=df.columns,index=df[-1*(n_days + 1):].index)
    upcoming_prediction.index=pd.to_datetime(upcoming_prediction.index)
    curr_seq = x

    for i in range(-1*n_days,0):
        up_pred = model.predict(curr_seq)
        upcoming_prediction.iloc[i] = up_pred
        curr_seq = np.append(curr_seq[0][1:],up_pred,axis=0)
        curr_seq = curr_seq.reshape(x[-1:].shape)

    upcoming_prediction[data.columns] = MMS.inverse_transform(upcoming_prediction[data.columns])

    return (upcoming_prediction)
```

```python
def calculate_rsi(data, window=14):
    delta = data['Adj Close'].diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

def calculate_macd(data, short_window=12, long_window=26, signal_window=9):
    exp1 = data['Adj Close'].ewm(span=short_window, adjust=False).mean()
    exp2 = data['Adj Close'].ewm(span=long_window, adjust=False).mean()
    macd = exp1 - exp2
    signal_line = macd.ewm(span=signal_window, adjust=False).mean()
    return macd, signal_line

def calculate_obv(data):
    obv = (data['Close'].diff().fillna(0) > 0).astype(int) * data['Volume']
    obv -= (data['Close'].diff().fillna(0) < 0).astype(int) * data['Volume']
    obv = obv.cumsum()
    return obv
```

```python
def recommend(stock_data,stock):
  stock_data['rsi']=calculate_rsi(stock_data)
  stock_data['macd'],stock_data['signal_line']=calculate_macd(stock_data)
  stock_data['obv']=calculate_obv(stock_data)

  if (stock_data['rsi'].iloc[-2]>70 and (stock_data['macd'].iloc[-2]-stock_data['signal_line'].iloc[-2])<0 and (stock_data['obv'].iloc[-2]<=stock_data['obv'].iloc[-3] or stock_data['obv'].iloc[-1]<stock_data['obv'].iloc[-3])):
    print(f'A trader should sell the {stock} stock as market conditions indicating the downfall')
  elif(30<stock_data['rsi'].iloc[-2]<70 and (stock_data['macd'].iloc[-2]-stock_data['signal_line'].iloc[-2])>0 and ((stock_data['obv'].iloc[-2]>stock_data['obv'].iloc[-3] and stock_data['obv'].iloc[-1]>stock_data['obv'].iloc[-3]) or stock_data['obv'].iloc[-2]>stock_data['obv'].iloc[-3])):
    print(f'A trader should buy the {stock} stock as market conditions indicating the increase in value.')
  else:
    print("hold your position")
```

```python
recommend(stock_data,stock)
```
